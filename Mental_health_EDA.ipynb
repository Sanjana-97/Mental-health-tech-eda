## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Write your code to make your dataset analysis ready.
# Clean Age column: Keep only values between 18 and 100
df = df[(df['Age'] >= 18) & (df['Age'] <= 100)]

# Clean Gender column: Normalize common entries
def clean_gender(g):
    g = str(g).strip().lower()
    if g in ['male', 'm', 'man', 'cis male', 'male-ish', 'maile', 'mal', 'msle', 'mail', 'make']:
        return 'Male'
    elif g in ['female', 'f', 'woman', 'cis female', 'femake', 'female ', 'cis-female/femme', 'femail']:
        return 'Female'
    elif 'trans' in g:
        return 'Trans'
    else:
        return 'Other'

df['Gender'] = df['Gender'].apply(clean_gender)

# Reset index after filtering
df.reset_index(drop=True, inplace=True)

df['self_employed'].fillna("Don't know", inplace=True)
df['work_interfere'].fillna("Don't know", inplace=True)
df.drop(columns=['comments'], inplace=True)

# Check results
print(df['Age'].describe())
print(df['Gender'].value_counts())
```

### What all manipulations have you done and insights you found?

#### Data Cleaning & Preparation

1. **Removed Duplicate Records**
   - No duplicate rows were found in the dataset.

2. **Identified and Handled Missing Values**
   - Used `df.isnull().sum()` and visualized missing data using a heatmap.
   - Filled missing values in:
     - `self_employed` → replaced with `"Don't know"` 
     - `work_interfere` → replaced with `"Don't know"` 
   - Dropped `comments` column due to excessive missing values and low relevance:  
     `df.drop(columns=['comments'], inplace=True)`
   - Kept `state` column for U.S.-specific analysis despite missing values.

3. **Cleaned the `Age` Column**
   - Removed unrealistic age entries (e.g., negative and extremely large values).
   - Kept only records where age is between **18 and 100**.
   - Insight: Majority of respondents are between **27 and 36 years old** (based on quartiles).

4. **Cleaned the `Gender` Column**
   - Standardized messy entries like "cis male", "Femake", "Mail", etc. into:
     - `Male`, `Female`, `Trans`, `Other`
   - Insight: 
     - **78.6%** identified as Male  
     - **19.7%** as Female  
     - Others (Trans/Other): ~1.7% combined

#### Dataset Understanding Highlights
